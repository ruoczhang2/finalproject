```python
import pandas as pd
```

```python
df1 = pd.read_csv('合并文书1.csv', encoding = 'utf8')
df1
```

```python
df2 = df1.dropna()
df2
```

```python
df2_accused = pd.DataFrame(df2[" accused"], columns=["accused"])
df2_accused.head()
```

```python
df2_accused = df2[" accused"]
df2_accused
```

```python
string=''.join(df2_accused)
search="文盲"
last_cur=0
count=0


while(1):
    where=string.find(search)
    if(not where==-1):
        print(last_cur+where)
        string=string[where+len(search):]
        last_cur=last_cur+where+len(search)
        count+=1
    else:
        break
    
print("Total：",count)
```
